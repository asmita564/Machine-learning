```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data into a DataFrame
data = pd.read_csv(r"C:\Users\pc\Documents\company data.csv")
data
```

```python
# Preprocess 'Revenue' column to remove non-numeric characters and convert to float
data['Revenue'] = data['Revenue'].replace('[^\d.]', '', regex=True).astype(float)

# Define a custom autopct function
def my_autopct(pct):
    return '{:.1f}%'.format(pct) if pct > 5 else ''

# Create a pie chart for Revenue
plt.figure(figsize=(12, 8))  # Increase the figure size
plt.pie(data['Revenue'], labels=None, autopct=my_autopct, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(data['Company name'], loc='best', fontsize='small', title='Company Names', title_fontsize='medium')
plt.title('Revenue Distribution of Companies', fontsize=16)  # Increase the title font size
plt.show()
```

```python
# Preprocess 'Revenue' column to remove non-numeric characters and convert to float
data['Revenue'] = data['Revenue'].replace('[^\d.]', '', regex=True).astype(float)

# Create a line plot for Revenue
plt.figure(figsize=(12, 8))  # Increase the figure size
for company in data['Company name'].unique():
    company_data = data[data['Company name'] == company]
    plt.plot(company_data['Founded year'], company_data['Revenue'], marker='o', label=company)

plt.xlabel('Founded Year')
plt.ylabel('Revenue')
plt.title('Revenue Trends of Companies')
plt.legend()
plt.grid(True)
plt.show()
```
